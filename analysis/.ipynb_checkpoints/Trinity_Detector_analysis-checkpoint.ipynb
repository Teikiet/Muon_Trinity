```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import os

# ── Load CSV ──────────────────────────────────────────────────────────────────
def load_scan(csv_path: str) -> pd.DataFrame:
    """Load the scan CSV and add derived columns."""
    if not os.path.exists(csv_path):
        raise FileNotFoundError(f"CSV not found: {csv_path}")
    
    df = pd.read_csv(csv_path)
    
    # Drop rows where file was not found
    n_missing = (df["file_found"] == 0).sum()
    if n_missing > 0:
        print(f"  WARNING: {n_missing} rows had file_found=0 and will be dropped.")
        #print out the parameters of the missing rows for debugging
        missing_rows = df[df["file_found"] == 0]
        print("  Missing rows parameters:")
        print(missing_rows[["tel_x", "tel_z", "zen", "az", "height"]])
    df = df[df["file_found"] == 1].copy()
    
    # Derived columns
    df["r"] = np.sqrt(df["tel_x"]**2 + df["tel_z"]**2).round(4)
    df["detected"] = (df["max_pe"] >= PE_THRESHOLD).astype(float)
    
    #print(f"Loaded {len(df)} valid rows from '{csv_path}'")
    #print(f"Columns: {list(df.columns)}")
    #print(df.describe())
    return df
def slant_depth(zem, obs_height, zenith_rad):
        theta = np.pi - zenith_rad
        beta = np.clip(np.arcsin((obs_height / zem) * np.sin(theta)), 0, np.pi/2)
        alpha = np.pi - theta - beta
        s = zem * np.sin(alpha) / np.sin(theta)
        return np.clip(s, 0, np.inf)

# ── Configuration ─────────────────────────────────────────────────────────────
PE_THRESHOLD = 20
N_BOOTSTRAP = 1000
CI_LEVEL = 0.99
E_mag = 5
SEEDS = np.arange(1, 27)  # Updated seeds based on recent edits
print(SEEDS)

# ── Merge all seeds ──────────────────────────────────────────────────────────
dfs = []
for seed in SEEDS:
    csv_path = f"/scratch/general/vast/u1520754/muon_sim_chain/Muon_pid13_E1e{E_mag}_R5/csv_output/scan_care_pid13_E1e{E_mag}_R5_y0_s{seed}.csv"
    try:
        df_tmp = load_scan(csv_path)
        df_tmp["seed"] = seed
        dfs.append(df_tmp)
        print(f"  Loaded seed {seed}: {len(df_tmp)} rows")
    except Exception as e:
        print(f"  ⚠ Seed {seed} failed: {e}")

df = pd.concat(dfs, ignore_index=True)
print(f"\nMerged: {len(df)} total rows from {len(dfs)} seeds")

# ── Derived columns ──────────────────────────────────────────────────────────
R = np.array(df["r"])
X = np.array(df["tel_x"])
Z = np.array(df["tel_z"])
H = np.array(df["height"])
Zenith = np.array(df["zen"])
Azimuth = np.array(df["az"])
Max_PE = np.array(df["max_pe"])
Time_at_Max_PE = np.array(df["time_at_max_pe_ns"])
Avg_PE = np.array(df["avg_pe"])
Total_PE = np.array(df["total_pe"])
Slant = slant_depth(H + 6371e3, 2944 + 6371e3, np.radians(Zenith))
cos_Zenith = np.cos(np.radians(Zenith))
cos_Azimuth = np.cos(np.radians(Azimuth - 180))
cos_Incidence = np.sqrt(1 - cos_Zenith**2 - cos_Azimuth**2)
Incidence = np.degrees(np.arccos(cos_Incidence))

variables = {
    "H": (H / 1000, "Production Height (km)", 5),
    "Slant": (Slant / 1000, "Slant Path (km)", 50),
    "Zenith": (Zenith, "Zenith (deg)", 0.3),
    "Azimuth": (Azimuth, "Azimuth (deg)", 0.3),
    "Incidence": (Incidence, "Incidence Angle (deg)", 0.3),
    #"R": (R, "R (m)", 0.2),
    #"X": (X, "X (m)", 0.2),
    #"Z": (Z, "Z (m)", 0.2),
    "Time at Max PE": (Time_at_Max_PE, "Time at Max PE (ns)", 10),
}

PE_threshold = PE_THRESHOLD
alpha = 1.0 - CI_LEVEL
lo_pct = 100 * alpha / 2
hi_pct = 100 * (1 - alpha / 2)

def bootstrap_trigger_rate(max_pe_in_bin, threshold, n_boot=N_BOOTSTRAP):
    n = len(max_pe_in_bin)
    if n == 0:
        return np.nan, np.nan, np.nan
    observed_rate = np.sum(max_pe_in_bin >= threshold) / n
    boot_rates = np.empty(n_boot)
    for b in range(n_boot):
        sample = np.random.choice(max_pe_in_bin, size=n, replace=True)
        boot_rates[b] = np.sum(sample >= threshold) / n
    ci_low = np.percentile(boot_rates, lo_pct)
    ci_high = np.percentile(boot_rates, hi_pct)
    return observed_rate, ci_low, ci_high

# ── First pass: collect all trigger counts to set shared color scale ─────────
all_counts = []
all_results = {}

for name, (var, xlabel, bin_size) in variables.items():
    var = np.array(var)
    mask = ~np.isnan(var)
    var = var[mask]
    max_pe_filtered = Max_PE[mask]

    if len(var) == 0:
        all_results[name] = None
        continue

    var_bins = np.arange(min(var) - bin_size, max(var) + 2 * bin_size, bin_size)
    trigger_rates = []
    ci_lows = []
    ci_highs = []
    bin_centers = []
    trigger_counts = []

    for i in range(len(var_bins) - 1):
        bin_mask = (var >= var_bins[i]) & (var < var_bins[i + 1])
        max_PE_in_bin = max_pe_filtered[bin_mask]
        if len(max_PE_in_bin) <= 1:
            continue

        rate, ci_lo, ci_hi = bootstrap_trigger_rate(max_PE_in_bin, PE_threshold)
        if rate == 0:
            continue
        trigger_rates.append(rate)
        ci_lows.append(ci_lo)
        ci_highs.append(ci_hi)
        bin_centers.append(var_bins[i] + bin_size / 2)
        trigger_counts.append(int(np.sum(max_PE_in_bin >= PE_threshold)))

    all_counts.extend(trigger_counts)
    all_results[name] = {
        "bin_centers": np.array(bin_centers),
        "trigger_rates": np.array(trigger_rates),
        "ci_lows": np.array(ci_lows),
        "ci_highs": np.array(ci_highs),
        "trigger_counts": np.array(trigger_counts),
    }

# ── Shared colormap normalization ────────────────────────────────────────────
if len(all_counts) > 0:
    vmin = min(all_counts)
    vmax = max(all_counts)
else:
    vmin, vmax = 0, 1

norm = mcolors.LogNorm(vmin=max(vmin, 1), vmax=max(vmax, 2))  # LogNorm for wide range
cmap = cm.viridis

# ── Plot ─────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(3, 2, figsize=(20, 20))
axes = axes.flatten()

for ax, (name, (_, xlabel, bin_size)) in zip(axes, variables.items()):
    res = all_results[name]
    if res is None:
        ax.set_title(f"{name} — no data")
        continue

    bin_centers = res["bin_centers"]
    trigger_rates = res["trigger_rates"]
    ci_lows = res["ci_lows"]
    ci_highs = res["ci_highs"]
    trigger_counts = res["trigger_counts"]

    yerr_lo = trigger_rates - ci_lows
    yerr_hi = ci_highs - trigger_rates

    # Plot error bars in gray
    ax.errorbar(
        bin_centers, trigger_rates,
        yerr=[yerr_lo, yerr_hi],
        fmt="none", ecolor="gray",
        elinewidth=1.2, capsize=3, capthick=1.0,
    )
    # Connect with dashed line
    ax.plot(bin_centers, trigger_rates, "--", color="gray", linewidth=0.8, alpha=0.5)

    # Scatter colored by trigger count
    sc = ax.scatter(
        bin_centers, trigger_rates,
        c=trigger_counts, cmap=cmap, norm=norm,
        s=50, edgecolors="black", linewidths=0.5, zorder=5,
    )

    ax.fill_between(bin_centers, ci_lows, ci_highs, alpha=0.12, color="coral")
    ax.set_xlabel(xlabel)
    ax.set_ylabel("Trigger Rate")
    ax.set_title(f"Trigger Rate vs {name}")
    ax.grid(True, alpha=0.3)

# ── Shared colorbar ──────────────────────────────────────────────────────────
fig.subplots_adjust(right=0.88)
cbar_ax = fig.add_axes([0.90, 0.08, 0.02, 0.84])
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, cax=cbar_ax)
cbar.set_label("Number of Trigger Events", fontsize=11)

plt.suptitle(
    f"Energy: 1e{E_mag} GeV,  PE threshold ≥ {PE_THRESHOLD},  Bootstrap 99% CI (N={N_BOOTSTRAP})",
    fontsize=14, x=0.44,
)
plt.show()
#2,7,8,10,21,22 incomplete data seed 
```

```python
import matplotlib.colors as mcolors

# --- Bin edges ---
slant_bins = np.linspace(Slant.min() / 1000, Slant.max() / 1000, 10)
inc_bins   = np.linspace(Incidence.min(), Incidence.max(), 10)

# --- Total counts per bin ---
total_counts, xedges, yedges = np.histogram2d(
    Slant / 1000, Incidence,
    bins=[slant_bins, inc_bins]
)

# --- Counts with Max_PE > 20 ---
mask = Max_PE > 20
counts_above, _, _ = np.histogram2d(
    Slant[mask] / 1000, Incidence[mask],
    bins=[slant_bins, inc_bins]
)
total_counts,_, _ = np.histogram2d(
    Slant / 1000, Incidence,
    bins=[slant_bins, inc_bins]
)
print("Total counts per bin (min and max):",np.min(counts_above), np.max(counts_above), np.sum(counts_above))
print("Total counts per bin (min and max):",np.min(total_counts), np.max(total_counts), np.sum(total_counts))
print("Average trigger rate", np.sum(counts_above) / np.sum(total_counts))
# --- Ratio: NaN where no events OR ratio is zero (log can't handle 0) ---
with np.errstate(invalid='ignore', divide='ignore'):
    ratio = np.where(total_counts > 0, counts_above / total_counts, np.nan)
print("max trigger rate in any bin:", np.nanmax(ratio))
# Set exact zeros to NaN so log scale doesn't break
ratio[ratio == 0] = np.nan

# --- Log norm: vmin is your effective minimum (e.g. 1 event out of ~100) ---
log_norm = mcolors.LogNorm(vmin=np.nanmin(ratio), vmax=None)

# --- Plot ---
fig, ax = plt.subplots(figsize=(7, 5))

im = ax.pcolormesh(
    xedges, yedges,
    ratio.T,
    cmap="viridis",
    norm=log_norm
)

cbar = fig.colorbar(im, ax=ax)
cbar.set_label("Ratio of events with Max PE > 20", fontsize=12)

# Optional: format colorbar ticks as percentages
#cbar.set_ticks([1e-4, 1e-1, 1])
#cbar.set_ticklabels(["1%", "10%", "100%"])

ax.set_xlabel("Slant Depth (km)", fontsize=13)
ax.set_ylabel("Incidence Angle (deg)", fontsize=13)
ax.set_title(
    "Fraction of Events with Max PE >= 20\nper (Slant Depth, Incidence Angle) Bin",
    fontsize=14
)
ax.grid(True, alpha=0.3, color='white', linewidth=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import os
import sys
import math

# ── Configuration ─────────────────────────────────────────────────────────────
PE_THRESHOLD = 20
N_BOOTSTRAP = 1000
CI_LEVEL = 0.99
SEEDS = [1]
PID = 13
RADIUS = 5
BASE_DIR = "/scratch/general/vast/u1520754/muon_sim_chain_tree"

# ── Get energy grid from MCEq ────────────────────────────────────────────────
print("Generating energy grid from MCEq...")
os.environ["MCEQ_LOG_LEVEL"] = "50"
import logging
logging.disable(logging.CRITICAL)
import MCEq.config as config
config.kernel_config = 'MKL'
config.e_min = 10000
config.integrator = 'euler'
old_stdout = sys.stdout
sys.stdout = open(os.devnull, 'w')
from MCEq.core import MCEqRun
import crflux.models as pm
mceq = MCEqRun(
    interaction_model='SIBYLL23C',
    primary_model=(pm.GlobalSplineFitBeta, None),
    theta_deg=0.,
    density_model=("CORSIKA", ('USStd', None)),
)
sys.stdout = old_stdout
logging.disable(logging.NOTSET)

E = mceq.e_grid
E = E[(E >= 1e4) & (E <= 1e6)]

def to_1e(val):
    exp = int(math.log10(val))
    coeff = val / 10**exp
    return f"{coeff:g}e{exp}"

energy_strings = [to_1e(e) for e in E]
print(f"Found {len(energy_strings)} energies from MCEq: {energy_strings[:5]} ... {energy_strings[-3:]}")

# ── Helper functions ─────────────────────────────────────────────────────────
def load_scan(csv_path: str) -> pd.DataFrame:
    if not os.path.exists(csv_path):
        raise FileNotFoundError(f"CSV not found: {csv_path}")
    df = pd.read_csv(csv_path)
    n_missing = (df["file_found"] == 0).sum()
    if n_missing > 0:
        print(f"  WARNING: {n_missing} rows had file_found=0 and will be dropped.")
    df = df[df["file_found"] == 1].copy()
    df["r"] = np.sqrt(df["tel_x"]**2 + df["tel_z"]**2).round(4)
    df["detected"] = (df["max_pe"] >= PE_THRESHOLD).astype(float)
    return df

def slant_depth(zem, obs_height, zenith_rad):
    theta = np.pi - zenith_rad
    beta = np.clip(np.arcsin((obs_height / zem) * np.sin(theta)), 0, np.pi/2)
    alpha = np.pi - theta - beta
    s = zem * np.sin(alpha) / np.sin(theta)
    return np.clip(s, 0, np.inf)

def bootstrap_trigger_rate(max_pe_in_bin, threshold, n_boot=N_BOOTSTRAP):
    alpha_ci = 1.0 - CI_LEVEL
    lo_pct = 100 * alpha_ci / 2
    hi_pct = 100 * (1 - alpha_ci / 2)
    n = len(max_pe_in_bin)
    if n == 0:
        return np.nan, np.nan, np.nan
    observed_rate = np.sum(max_pe_in_bin >= threshold) / n
    boot_rates = np.empty(n_boot)
    for b in range(n_boot):
        sample = np.random.choice(max_pe_in_bin, size=n, replace=True)
        boot_rates[b] = np.sum(sample >= threshold) / n
    ci_low = np.percentile(boot_rates, lo_pct)
    ci_high = np.percentile(boot_rates, hi_pct)
    return observed_rate, ci_low, ci_high

# ── Load and merge all energies and seeds ────────────────────────────────────
dfs = []
missing_energies = []
for energy_str in energy_strings:
    for seed in SEEDS:
        csv_path = (f"{BASE_DIR}/Muon_pid{PID}_E{energy_str}_R{RADIUS}/"
                    f"csv_output/scan_care_pid{PID}_E{energy_str}_R{RADIUS}_y0_s{seed}.csv")
        try:
            df_tmp = load_scan(csv_path)
            df_tmp["seed"] = seed
            df_tmp["energy_string"] = energy_str
            df_tmp["energy_GeV"] = float(energy_str)
            dfs.append(df_tmp)
            print(f"  Loaded E={energy_str} seed={seed}: {len(df_tmp)} rows")
        except Exception as e:
            missing_energies.append(energy_str)
            print(f"  ⚠ E={energy_str} seed={seed}: {e}")

if not dfs:
    print("No data loaded!")
    sys.exit(1)

df = pd.concat(dfs, ignore_index=True)
print(f"\nMerged: {len(df)} total rows from {len(dfs)} file(s)")
print(f"Energy range: {df['energy_GeV'].min():.1f} – {df['energy_GeV'].max():.1f} GeV")
if missing_energies:
    print(f"Missing energies: {sorted(set(missing_energies))}")

# ── Derived columns ──────────────────────────────────────────────────────────
R = np.array(df["r"])
X = np.array(df["tel_x"])
Z = np.array(df["tel_z"])
H = np.array(df["height"])
Zenith = np.array(df["zen"])
Azimuth = np.array(df["az"])
Max_PE = np.array(df["max_pe"])
Time_at_Max_PE = np.array(df["time_at_max_pe_ns"])
Avg_PE = np.array(df["avg_pe"])
Total_PE = np.array(df["total_pe"])
Energy_GeV = np.array(df["energy_GeV"])
Slant = slant_depth(H + 6371e3, 2944 + 6371e3, np.radians(Zenith))
cos_Zenith = np.cos(np.radians(Zenith))
cos_Azimuth = np.cos(np.radians(Azimuth - 180))
cos_Incidence = np.sqrt(1 - cos_Zenith**2 - cos_Azimuth**2)
Incidence = np.degrees(np.arccos(cos_Incidence))

variables = {
    "Energy": (Energy_GeV / 1e3, "Energy (TeV)", 10),
    "H": (H / 1000, "Production Height (km)", 5),
    "Slant": (Slant / 1000, "Slant Path (km)", 50),
    "Zenith": (Zenith, "Zenith (deg)", 0.3),
    "Azimuth": (Azimuth, "Azimuth (deg)", 0.3),
    "Time at Max PE": (Time_at_Max_PE, "Time at Max PE (ns)", 10),
    "Incidence": (Incidence, "Incidence (deg)", 0.3),
}

# ── First pass: collect results ──────────────────────────────────────────────
all_counts = []
all_results = {}

for name, (var, xlabel, bin_size) in variables.items():
    var = np.array(var)
    mask = ~np.isnan(var)
    var = var[mask]
    max_pe_filtered = Max_PE[mask]

    if len(var) == 0:
        all_results[name] = None
        continue

    if name == "Energy":
        var_bins = np.logspace(np.log10(var.min() * 0.9), np.log10(var.max() * 1.1), 20)
    else:
        var_bins = np.arange(min(var) - bin_size, max(var) + 2 * bin_size, bin_size)

    trigger_rates, ci_lows, ci_highs, bin_centers, trigger_counts = [], [], [], [], []

    for i in range(len(var_bins) - 1):
        bin_mask = (var >= var_bins[i]) & (var < var_bins[i + 1])
        max_PE_in_bin = max_pe_filtered[bin_mask]
        if len(max_PE_in_bin) <= 1:
            continue
        rate, ci_lo, ci_hi = bootstrap_trigger_rate(max_PE_in_bin, PE_THRESHOLD)
        if rate == 0:
            continue
        trigger_rates.append(rate)
        ci_lows.append(ci_lo)
        ci_highs.append(ci_hi)
        if name == "Energy":
            bin_centers.append(np.sqrt(var_bins[i] * var_bins[i + 1]))
        else:
            bin_centers.append(var_bins[i] + bin_size / 2)
        trigger_counts.append(int(np.sum(max_PE_in_bin >= PE_THRESHOLD)))

    all_counts.extend(trigger_counts)
    all_results[name] = {
        "bin_centers": np.array(bin_centers),
        "trigger_rates": np.array(trigger_rates),
        "ci_lows": np.array(ci_lows),
        "ci_highs": np.array(ci_highs),
        "trigger_counts": np.array(trigger_counts),
    }

# ── Shared colormap normalization ────────────────────────────────────────────
if len(all_counts) > 0:
    vmin, vmax = min(all_counts), max(all_counts)
else:
    vmin, vmax = 0, 1
norm = mcolors.LogNorm(vmin=max(vmin, 1), vmax=max(vmax, 2))
cmap = cm.viridis

# ── Plot ─────────────────────────────────────────────────────────────────────
n_vars = len(variables)
ncols = 2
nrows = (n_vars + 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(20, 7 * nrows))
axes = axes.flatten()

for ax, (name, (_, xlabel, bin_size)) in zip(axes, variables.items()):
    res = all_results[name]
    if res is None:
        ax.set_title(f"{name} — no data")
        continue

    bin_centers = res["bin_centers"]
    trigger_rates = res["trigger_rates"]
    ci_lows = res["ci_lows"]
    ci_highs = res["ci_highs"]
    trigger_counts = res["trigger_counts"]

    yerr_lo = trigger_rates - ci_lows
    yerr_hi = ci_highs - trigger_rates

    ax.errorbar(bin_centers, trigger_rates, yerr=[yerr_lo, yerr_hi],
                fmt="none", ecolor="gray", elinewidth=1.2, capsize=3, capthick=1.0)
    ax.plot(bin_centers, trigger_rates, "--", color="gray", linewidth=0.8, alpha=0.5)
    sc = ax.scatter(bin_centers, trigger_rates, c=trigger_counts, cmap=cmap, norm=norm,
                    s=50, edgecolors="black", linewidths=0.5, zorder=5)
    ax.fill_between(bin_centers, ci_lows, ci_highs, alpha=0.12, color="coral")

    ax.set_xlabel(xlabel)
    ax.set_ylabel("Trigger Rate")
    ax.set_title(f"Trigger Rate vs {name}")
    ax.grid(True, alpha=0.3)

    if name == "Energy":
        ax.set_xscale("log")

for i in range(len(variables), len(axes)):
    axes[i].set_visible(False)

# ── Shared colorbar ──────────────────────────────────────────────────────────
fig.subplots_adjust(right=0.88)
cbar_ax = fig.add_axes([0.90, 0.08, 0.02, 0.84])
sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, cax=cbar_ax)
cbar.set_label("Number of Trigger Events", fontsize=11)

energy_range = f"{df['energy_GeV'].min():.0f}–{df['energy_GeV'].max():.0f} GeV"
plt.suptitle(
    f"All Energies ({energy_range}),  PE threshold ≥ {PE_THRESHOLD},  "
    f"Seeds: {SEEDS},  Bootstrap {int(CI_LEVEL*100)}% CI (N={N_BOOTSTRAP})",
    fontsize=14, x=0.44,
)
plt.show()
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, average_precision_score, classification_report
from sklearn.calibration import CalibratedClassifierCV
import joblib, warnings
warnings.filterwarnings("ignore")

print("\n── ML Model: Trigger Rate Interpolation ─────────────────────────────")

# ── Features ──────────────────────────────────────────────────────────────────
log_E   = np.log10(Energy_GeV)
H_km    = H / 1000.0
zen_rad = np.radians(Zenith)
az_rad  = np.radians(Azimuth)

features = np.column_stack([
    log_E,
    H_km,
    np.cos(zen_rad),
    np.sin(zen_rad),
    np.cos(az_rad),
    np.sin(az_rad),
])

target = (Max_PE >= PE_THRESHOLD).astype(int)

valid = np.all(np.isfinite(features), axis=1) & np.isfinite(target)
X_ml  = features[valid]
y_ml  = target[valid]

n_pos = y_ml.sum()
n_neg = (y_ml == 0).sum()
scale_pos_weight = n_neg / max(n_pos, 1)   # class imbalance ratio

print(f"Dataset : {X_ml.shape[0]} rows")
print(f"Positive: {n_pos}  ({100*n_pos/len(y_ml):.3f}%)")
print(f"Negative: {n_neg}")
print(f"Class weight ratio: {scale_pos_weight:.1f}x")

# ── Oversample positives (SMOTE-free: repeat positives) ───────────────────────
X_pos   = X_ml[y_ml == 1]
y_pos   = y_ml[y_ml == 1]
X_neg   = X_ml[y_ml == 0]
y_neg   = y_ml[y_ml == 0]

# Upsample positives to 20% of negatives
n_upsample = min(int(0.20 * len(X_neg)), len(X_pos) * 50)
rng = np.random.default_rng(42)
idx = rng.choice(len(X_pos), size=n_upsample, replace=True)
X_bal = np.vstack([X_neg, X_pos[idx]])
y_bal = np.concatenate([y_neg, y_pos[idx]])

# Shuffle
perm  = rng.permutation(len(X_bal))
X_bal = X_bal[perm]
y_bal = y_bal[perm]

print(f"\nBalanced dataset: {len(X_bal)} rows  "
      f"(pos={y_bal.sum()}, {100*y_bal.mean():.1f}%)")

# ── Train / test split ────────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.2, random_state=42, stratify=y_bal
)

scaler  = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

# ── Gradient Boosting Classifier ──────────────────────────────────────────────
clf = GradientBoostingClassifier(
    n_estimators=400,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    min_samples_leaf=10,
    random_state=42,
)
clf.fit(X_train_s, y_train)

y_prob_test  = clf.predict_proba(X_test_s)[:, 1]
y_prob_train = clf.predict_proba(X_train_s)[:, 1]

roc_auc  = roc_auc_score(y_test, y_prob_test)
pr_auc   = average_precision_score(y_test, y_prob_test)

print(f"\nROC-AUC  : {roc_auc:.4f}   (1.0 = perfect, 0.5 = random)")
print(f"PR-AUC   : {pr_auc:.4f}   (meaningful for imbalanced data)")

# ── Feature importance ────────────────────────────────────────────────────────
feature_names = ["log10(E/GeV)", "H (km)", "cos(zen)", "sin(zen)", "cos(az)", "sin(az)"]
importances   = clf.feature_importances_

fig_imp, ax_imp = plt.subplots(figsize=(7, 4))
bars = ax_imp.barh(feature_names, importances, color="steelblue", edgecolor="black")
ax_imp.set_xlabel("Feature Importance", fontsize=12)
ax_imp.set_title("Feature Importances — Trigger Rate Classifier", fontsize=13)
ax_imp.bar_label(bars, fmt="%.3f", padding=3)
ax_imp.grid(True, axis="x", alpha=0.3)
plt.tight_layout()
plt.show()

# ── Calibrated trigger rate vs Energy (on ORIGINAL unbalanced data) ───────────
X_orig_s   = scaler.transform(features[valid])
prob_orig   = clf.predict_proba(X_orig_s)[:, 1]

E_valid    = Energy_GeV[valid]
e_bins     = np.logspace(np.log10(E_valid.min()*0.9), np.log10(E_valid.max()*1.1), 20)
e_centers, rate_true, rate_pred = [], [], []

for i in range(len(e_bins) - 1):
    mask_e = (E_valid >= e_bins[i]) & (E_valid < e_bins[i+1])
    if mask_e.sum() < 5:
        continue
    e_centers.append(np.sqrt(e_bins[i] * e_bins[i+1]))
    rate_true.append(y_ml[mask_e].mean())
    rate_pred.append(prob_orig[mask_e].mean())

e_centers = np.array(e_centers)
rate_true = np.array(rate_true)
rate_pred = np.array(rate_pred)

fig3, ax3 = plt.subplots(figsize=(8, 5))
ax3.plot(e_centers/1e3, rate_true, "o-",  color="steelblue", label="Observed rate",    lw=2)
ax3.plot(e_centers/1e3, rate_pred, "s--", color="tomato",    label="ML predicted rate", lw=2)
ax3.set_xscale("log")
ax3.set_xlabel("Energy (TeV)", fontsize=13)
ax3.set_ylabel("Trigger Rate", fontsize=13)
ax3.set_title("Observed vs Predicted Trigger Rate vs Energy", fontsize=14)
ax3.legend(fontsize=11)
ax3.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ── 2D slice: Energy × Zenith ────────────────────────────────────────────────
E_grid   = np.logspace(np.log10(E_valid.min()), np.log10(E_valid.max()), 60)
zen_grid = np.linspace(Zenith.min(), Zenith.max(), 60)
EE, ZZ   = np.meshgrid(E_grid, zen_grid)
H_med    = np.median(H_km)
az_med   = np.radians(np.median(Azimuth))

grid_feat = np.column_stack([
    np.log10(EE.ravel()),
    np.full(EE.size, H_med),
    np.cos(np.radians(ZZ.ravel())),
    np.sin(np.radians(ZZ.ravel())),
    np.full(EE.size, np.cos(az_med)),
    np.full(EE.size, np.sin(az_med)),
])
grid_pred = clf.predict_proba(scaler.transform(grid_feat))[:, 1].reshape(EE.shape)

fig4, ax4 = plt.subplots(figsize=(9, 5))
pcm = ax4.pcolormesh(E_grid/1e3, zen_grid, grid_pred, cmap="viridis", shading="auto")
fig4.colorbar(pcm, ax=ax4, label="Predicted Trigger Rate")
ax4.set_xscale("log")
ax4.set_xlabel("Energy (TeV)", fontsize=13)
ax4.set_ylabel("Zenith (°)", fontsize=13)
ax4.set_title(f"Predicted Trigger Rate — Energy × Zenith\n"
              f"(H={H_med:.1f} km, Az={np.degrees(az_med):.1f}° [medians])", fontsize=13)
plt.tight_layout()
plt.show()

joblib.dump({"model": clf, "scaler": scaler}, "trigger_rate_clf.pkl")
print("\nModel saved → trigger_rate_clf.pkl")
```

```python
# ── Validation on ORIGINAL unbalanced held-out set ───────────────────────────
print("\n── Validation on original unbalanced data ───────────────────────────")

# Split ORIGINAL (unbalanced) data
X_orig_all = features[valid]
y_orig_all  = y_ml  # original 0/1, unbalanced

X_tr_orig, X_te_orig, y_tr_orig, y_te_orig = train_test_split(
    X_orig_all, y_orig_all, test_size=0.2, random_state=42, stratify=y_orig_all
)
X_te_orig_s = scaler.transform(X_te_orig)
prob_orig_test = clf.predict_proba(X_te_orig_s)[:, 1]

roc_orig = roc_auc_score(y_te_orig, prob_orig_test)
pr_orig  = average_precision_score(y_te_orig, prob_orig_test)

print(f"On held-out ORIGINAL data (n={len(y_te_orig)}, pos={y_te_orig.sum()}):")
print(f"  ROC-AUC : {roc_orig:.4f}")
print(f"  PR-AUC  : {pr_orig:.4f}")

# ── Threshold sweep: find best operating point ────────────────────────────────
from sklearn.metrics import precision_recall_curve, f1_score

precision_arr, recall_arr, thresholds = precision_recall_curve(y_te_orig, prob_orig_test)
f1_arr = np.where(
    (precision_arr + recall_arr) > 0,
    2 * precision_arr * recall_arr / (precision_arr + recall_arr),
    0
)
best_idx   = np.argmax(f1_arr)
best_thresh = thresholds[best_idx] if best_idx < len(thresholds) else thresholds[-1]
best_f1     = f1_arr[best_idx]

print(f"\nBest threshold : {best_thresh:.4f}")
print(f"Best F1        : {best_f1:.4f}")
print(f"  Precision    : {precision_arr[best_idx]:.4f}")
print(f"  Recall       : {recall_arr[best_idx]:.4f}")

y_pred_best = (prob_orig_test >= best_thresh).astype(int)
print(f"\nClassification report at best threshold:")
print(classification_report(y_te_orig, y_pred_best, digits=4))

# ── Precision-Recall curve ────────────────────────────────────────────────────
fig_pr, ax_pr = plt.subplots(figsize=(7, 5))
ax_pr.plot(recall_arr, precision_arr, color="steelblue", lw=2, label=f"PR curve (AUC={pr_orig:.4f})")
ax_pr.axvline(recall_arr[best_idx], color="tomato", linestyle="--", alpha=0.7)
ax_pr.axhline(precision_arr[best_idx], color="tomato", linestyle="--", alpha=0.7,
              label=f"Best threshold={best_thresh:.3f}\nF1={best_f1:.4f}")
ax_pr.scatter([recall_arr[best_idx]], [precision_arr[best_idx]],
              color="tomato", s=80, zorder=5)
ax_pr.set_xlabel("Recall", fontsize=13)
ax_pr.set_ylabel("Precision", fontsize=13)
ax_pr.set_title("Precision-Recall Curve (original unbalanced test set)", fontsize=13)
ax_pr.legend(fontsize=11)
ax_pr.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ── Reliability / calibration plot ───────────────────────────────────────────
from sklearn.calibration import calibration_curve

# Need enough positives: use ~8 bins max given only ~13 positives in test
n_bins_cal = min(8, max(2, y_te_orig.sum() // 2))
frac_pos, mean_pred = calibration_curve(y_te_orig, prob_orig_test,
                                         n_bins=n_bins_cal, strategy="quantile")

fig_cal, ax_cal = plt.subplots(figsize=(6, 5))
ax_cal.plot([0, 1], [0, 1], "k--", label="Perfect calibration")
ax_cal.plot(mean_pred, frac_pos, "o-", color="steelblue", lw=2, label="Model")
ax_cal.set_xlabel("Mean predicted probability", fontsize=13)
ax_cal.set_ylabel("Fraction of positives", fontsize=13)
ax_cal.set_title("Calibration Plot (original test set)", fontsize=13)
ax_cal.legend(fontsize=11)
ax_cal.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ── Learning curve: check overfitting ────────────────────────────────────────
from sklearn.model_selection import learning_curve

print("\nComputing learning curve (may take ~1 min)...")
train_sizes, train_scores, val_scores = learning_curve(
    GradientBoostingClassifier(
        n_estimators=200, max_depth=5, learning_rate=0.05,
        subsample=0.8, min_samples_leaf=10, random_state=42
    ),
    X_bal, y_bal,
    train_sizes=np.linspace(0.1, 1.0, 8),
    scoring="roc_auc",
    cv=3,
    n_jobs=-1,
)

fig_lc, ax_lc = plt.subplots(figsize=(7, 5))
ax_lc.plot(train_sizes, train_scores.mean(axis=1), "o-", color="steelblue", label="Train ROC-AUC")
ax_lc.fill_between(train_sizes,
                    train_scores.mean(1) - train_scores.std(1),
                    train_scores.mean(1) + train_scores.std(1), alpha=0.15, color="steelblue")
ax_lc.plot(train_sizes, val_scores.mean(axis=1),  "o-", color="tomato",    label="CV ROC-AUC")
ax_lc.fill_between(train_sizes,
                    val_scores.mean(1) - val_scores.std(1),
                    val_scores.mean(1) + val_scores.std(1), alpha=0.15, color="tomato")
ax_lc.set_xlabel("Training set size", fontsize=13)
ax_lc.set_ylabel("ROC-AUC", fontsize=13)
ax_lc.set_title("Learning Curve — Gradient Boosting Classifier", fontsize=13)
ax_lc.legend(fontsize=11)
ax_lc.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"\nSummary:")
print(f"  Balanced test  ROC-AUC : {roc_auc:.4f}")
print(f"  Original test  ROC-AUC : {roc_orig:.4f}")
print(f"  Original test  PR-AUC  : {pr_orig:.4f}")
print(f"  Best F1 threshold       : {best_thresh:.4f}")
if roc_orig > 0.85:
    print("  ✓ Model generalises well to original unbalanced data")
else:
    print("  ✗ Overfitting suspected — model memorised upsampled duplicates")
```

```python
# ── Physics interpretation (no SHAP required) ─────────────────────────────────
print("\n── Physics Interpretation ───────────────────────────────────────────")

from sklearn.inspection import permutation_importance, PartialDependenceDisplay

feature_names = ["log10(E/GeV)", "H (km)", "cos(zen)", "sin(zen)", "cos(az)", "sin(az)"]

# ── 1. Built-in feature importance (from GBC) ────────────────────────────────
fi_builtin = clf.feature_importances_
sorted_idx = np.argsort(fi_builtin)[::-1]

print("\nBuilt-in feature importance (mean decrease in impurity):")
print(f"  {'Feature':<18} {'Importance':>12}")
print("  " + "-" * 32)
for idx in sorted_idx:
    print(f"  {feature_names[idx]:<18} {fi_builtin[idx]:>12.5f}")

# ── 2. Permutation importance on original unbalanced test set ─────────────────
print("\nComputing permutation importance (may take ~30 s)...")
perm = permutation_importance(
    clf, X_te_orig_s, y_te_orig,
    scoring="roc_auc",
    n_repeats=20,
    random_state=42,
    n_jobs=-1,
)
perm_sorted = np.argsort(perm.importances_mean)[::-1]

print("\nPermutation importance (drop in ROC-AUC when feature is shuffled):")
print(f"  {'Feature':<18} {'Mean drop':>12}  {'± Std':>10}")
print("  " + "-" * 44)
for idx in perm_sorted:
    print(f"  {feature_names[idx]:<18} {perm.importances_mean[idx]:>12.5f}"
          f"  ±{perm.importances_std[idx]:>9.5f}")

# ── 3. Plot: both importances side by side ───────────────────────────────────
fig_fi, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# Built-in
colors = ["steelblue" if i == sorted_idx[0] else "lightsteelblue" for i in range(len(feature_names))]
ax1.barh([feature_names[i] for i in sorted_idx[::-1]],
         fi_builtin[sorted_idx[::-1]],
         color=colors[::-1])
ax1.set_xlabel("Mean Decrease in Impurity", fontsize=12)
ax1.set_title("Built-in Feature Importance", fontsize=13)
ax1.grid(True, alpha=0.3, axis="x")

# Permutation
perm_means = perm.importances_mean[perm_sorted[::-1]]
perm_stds  = perm.importances_std[perm_sorted[::-1]]
ax2.barh([feature_names[i] for i in perm_sorted[::-1]],
         perm_means,
         xerr=perm_stds,
         color="tomato", alpha=0.8, capsize=4)
ax2.set_xlabel("Mean ROC-AUC Drop", fontsize=12)
ax2.set_title("Permutation Importance (unbalanced test set)", fontsize=13)
ax2.grid(True, alpha=0.3, axis="x")

plt.suptitle("Feature Importance Comparison", fontsize=14)
plt.tight_layout()
plt.show()

# ── 4. Partial dependence plots (sklearn built-in) ───────────────────────────
print("\nComputing partial dependence plots...")

fig_pd, axes_pd = plt.subplots(2, 3, figsize=(15, 9))

PartialDependenceDisplay.from_estimator(
    clf,
    X_te_orig_s,
    features=[0, 1, 2, 3, 4, 5],
    feature_names=feature_names,
    kind="average",           # pure PD (fast); use "both" to overlay ICE curves
    ax=axes_pd.ravel(),
    line_kw={"color": "steelblue", "lw": 2},
)
plt.suptitle("Partial Dependence Plots\n(marginalised over all other features)", fontsize=14)
plt.tight_layout()
plt.show()

# ── 5. 2D partial dependence: E×zenith and E×height ─────────────────────────
print("Computing 2D partial dependence...")

fig_2d, axes_2d = plt.subplots(1, 2, figsize=(13, 5))

PartialDependenceDisplay.from_estimator(
    clf,
    X_te_orig_s,
    features=[(0, 2), (0, 1)],          # (log_E, cos_zen), (log_E, H)
    feature_names=feature_names,
    kind="average",
    ax=axes_2d,
)
axes_2d[0].set_title("2D PD: log10(E) × cos(zenith)", fontsize=12)
axes_2d[1].set_title("2D PD: log10(E) × H (km)",      fontsize=12)
plt.suptitle("2D Partial Dependence — Trigger Rate", fontsize=14)
plt.tight_layout()
plt.show()

# ── 6. Physics summary ───────────────────────────────────────────────────────
top_perm = feature_names[perm_sorted[0]]
top_btin = feature_names[sorted_idx[0]]
print(f"""
Physics summary:
  Most important (permutation) : {top_perm}
  Most important (impurity)    : {top_btin}
  Note: permutation importance on the held-out unbalanced test set
        is more reliable than impurity for imbalanced problems.
  Best prediction threshold    : {best_thresh:.4f}
  Positive class prevalence    : {100*y_ml.mean():.3f}%
""")
```
